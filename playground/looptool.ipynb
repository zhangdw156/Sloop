```python
data_path="/dfs/data/datasets/LoopTool-23k/LoopTool_grpo_training_data.json"
```

```python
import pandas as pd
```

```python
df=pd.read_json(data_path,orient="record")
```

```python
df.info()
```

```python
# 计算 instruction + input + output 的总长度的最大值
max_len = (df['instruction'].str.len() + df['input'].str.len() + df['output'].str.len()).max()

print(f"最长的数据长度是: {max_len}")
```

```python
# 创建一个临时列存放总长度
df['total_len'] = df['instruction'].str.len() + df['input'].str.len() + df['output'].str.len()

# 找到最长的那一行
longest_row = df.loc[df['total_len'].idxmax()]

print(f"最大长度: {longest_row['total_len']}")
print("--- 数据内容预览 ---")
# print(longest_row)
```

```python
mean_len = (df['instruction'].str.len() + df['input'].str.len() + df['output'].str.len()).mean()

print(f"平均的数据长度是: {mean_len}")
```
